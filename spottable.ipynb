```python
import json
from datetime import datetime, timezone
curr = datetime.now(timezone.utc)
format = "%B %Y"
format1 = "%Y-%m-%d"
format2 = "%Y-%m-%dT%H:%M:%S.%f%z"
format3 = "%Y"
format4 = "%Y-%m"
format5 = "%b %Y"
user_gap = []
for k in range(21):
 with open('talents_sample.json') as json_file:
   data = json.load(json_file)[k]
 print("Employee",k+1)   
#print(type(data['Experience']))
 if data['Experience']:
  expe=[]
  dcd=0
  sd = []
  ed = []
  for i in data['Experience']: 
    try:
      datetime.strptime(i['startDate'], format)
      datem = datetime.strptime(i['startDate'], "%B %Y")
      if i['endDate']=="current" or i['endDate']=="Present":
        d = curr - datem
      else: 
        if i['endDate']:
            datetime.strptime(i['endDate'], format)
            datemn = datetime.strptime(i['endDate'], "%B %Y")
            d = datemn - datem
        else:
            datt = datetime.now()
            d = datt - datem
      expe.append(d.days)
      ed.append(datemn.year)
      sd.append(datem.year)
    except ValueError:
       try:
         datetime.strptime(i['startDate'], format1)
         datem = datetime.strptime(i['startDate'], "%Y-%m-%d")
         if i['endDate']=="current" or i['endDate']=="Present":
          d = curr - datem
         else: 
            if i['endDate']:
              datetime.strptime(i['endDate'], format1)
              datemn = datetime.strptime(i['endDate'], "%Y-%m-%d")
              d = datemn - datem
            else:
               datt = datetime.now()
               d = datt - datem
         expe.append(d.days)
         ed.append(datemn.year)
         sd.append(datem.year)
       except ValueError:
         try:
           datetime.strptime(i['startDate'], format2)
           datem = datetime.strptime(i['startDate'], "%Y-%m-%dT%H:%M:%S.%f%z")
           if i['endDate']=="current" or i['endDate']=="Present":
                  d = curr - datem
           else: 
                if i['endDate']:
                   datetime.strptime(i['endDate'], format2)
                   datemn = datetime.strptime(i['endDate'],"%Y-%m-%dT%H:%M:%S.%f%z")
                   d = datemn - datem
                else:
                   d = curr - datem            
           expe.append(d.days)
           ed.append(datemn.year)
           sd.append(datem.year)        
         except ValueError:
           try:
            datetime.strptime(i['startDate'], format3)
            datem = datetime.strptime(i['startDate'], "%Y")
            if i['endDate']=="current" or i['endDate']=="Present":
                  d = curr - datem
            else: 
                if i['endDate']:
                   datetime.strptime(i['endDate'], format3)
                   datemn = datetime.strptime(i['endDate'], "%Y")
                   d = datemn - datem
                else:
                    datt = datetime.now()
                    d = datt - datem             
            expe.append(d.days)
            ed.append(datemn.year)
            sd.append(datem.year)
           except ValueError:
             try:
              datetime.strptime(i['startDate'], format4)
              datem = datetime.strptime(i['startDate'], "%Y-%m")
              if i['endDate']=="current" or i['endDate']=="Present":
                  datt = datetime.now()
                  d = datt - datem
              else: 
                if i['endDate']:
                   datetime.strptime(i['endDate'], format4)
                   datemn = datetime.strptime(i['endDate'], "%Y-%m")
                   d = datemn - datem
                else:
                   datt = datetime.now()
                   d = datt - datem          
              expe.append(d.days)
              ed.append(datemn.year)
              sd.append(datem.year)
             except ValueError:
               try:
                datetime.strptime(i['startDate'], format5)
                datem = datetime.strptime(i['startDate'], "%b %Y")
                if i['endDate']=="current" or i['endDate']=="Present":
                  datt = datetime.now()
                  d = datt - datem
                else: 
                 if i['endDate']:
                   datetime.strptime(i['endDate'], format5)
                   datemn = datetime.strptime(i['endDate'], "%b %Y")
                   d = datemn - datem
                 else:
                   d = curr - datem
                expe.append(d.days)    
                ed.append(datemn.year)
                sd.append(datem.year)
               except ValueError:
                dcd = 1
                
  if dcd == 0:     
   minn=min(expe)
   for z in range(len(expe)):
     if minn==expe[z]:
       ind=z
   print("Company with least Experience : ",data['Experience'][ind]['companyName'])
   mxnn=max(expe)
   for l in range(len(expe)):
    if mxnn==expe[l]:
      indc=l
   print("Company with most Experience : ",data['Experience'][indc]['companyName'])
 
   sm = sum(expe)
   yearss = sm//365
   monthss = (sm - yearss *365) // 30
   dayss = (sm - yearss * 365 - monthss*30)
   print("Total experience",yearss,"Year ,",monthss,"Months and ",dayss,"days")
  else:
    minn=min(expe)
    for z in range(len(expe)):
      if minn==expe[z]:
         ind=z+1
    print("Company with least Experience : ",data['Experience'][ind]['companyName'])
    mxnn=max(expe)
    for l in range(len(expe)):
     if mxnn==expe[l]:
      indc=l+1
    print("Company with most Experience : ",data['Experience'][indc]['companyName'])
 
    sm = sum(expe)
    yearss = sm//365
    monthss = (sm - yearss *365) // 30
    dayss = (sm - yearss * 365 - monthss*30)
    print("Total experience",yearss,"Year ,",monthss,"Months and ",dayss,"days")
  sd.sort()
  ed.sort() 
  count=0
  o = 0
  for k in range(len(sd)):
    try:
     if sd[k+1]!=ed[o]:
       count = count + 1
       break 
    except:
      print("invalid")
  if count!=0:
      user_gap.append(data['EmailIds'])

 
 else :
    print("No experience")
print ("Number of employess who had a gap ",len(user_gap))
print("Following is the list of Employees who had a gap")
for q in range(len(user_gap)):
    print(user_gap[q])
```
